```python
df_limpo = df.copy()

df_limpo = df_limpo.dropna(subset=['COMBUSTÍVEL (LITROS)']) # é a coluna com mais valores nulos e não vou usar a info
df_limpo = df_limpo.dropna(subset=['PASSAGEIROS PAGOS']) #vou usar esses dados, por isso não posso ter valores nulos. Como são poucos registros nulos, vou excluir
df_limpo = df_limpo.dropna(subset=['ASK']) #mesmo anterior
df_limpo = df_limpo.drop(columns=['HORAS VOADAS']) #nãousarei
df_limpo = df_limpo.drop(columns=['BAGAGEM (KG)']) #nãousarei
df_limpo = df_limpo[df_limpo['GRUPO DE VOO'] != 'IMPRODUTIVO']

```

```python
verificacao_natureza = df_limpo[df_limpo['AEROPORTO DE ORIGEM (UF)'].isnull()]['NATUREZA'].value_counts()

print("Natureza dos voos com UF nula:")
print(verificacao_natureza)
```

```python
verificacao_natureza = df_limpo[df_limpo['AEROPORTO DE DESTINO (UF)'].isnull()]['NATUREZA'].value_counts()

print("Natureza dos voos com UF nula:")
print(verificacao_natureza)
```

```python
df_limpo.isnull().sum()
```

```python
df_limpo['AEROPORTO DE ORIGEM (UF)'] = df_limpo['AEROPORTO DE ORIGEM (UF)'].fillna('EXTERIOR')
df_limpo['AEROPORTO DE ORIGEM (REGIÃO)'] = df_limpo['AEROPORTO DE ORIGEM (REGIÃO)'].fillna('EXTERIOR')

df_limpo['AEROPORTO DE DESTINO (UF)'] = df_limpo['AEROPORTO DE DESTINO (UF)'].fillna('EXTERIOR')
df_limpo['AEROPORTO DE DESTINO (REGIÃO)'] = df_limpo['AEROPORTO DE DESTINO (REGIÃO)'].fillna('EXTERIOR')

df_limpo['PASSAGEIROS PAGOS'] = df_limpo['PASSAGEIROS PAGOS'].astype('int64')
df_limpo['PASSAGEIROS GRÁTIS'] = df_limpo['PASSAGEIROS GRÁTIS'].astype('int64')
df_limpo['DECOLAGENS'] = df_limpo['DECOLAGENS'].astype('int64')
df_limpo['ASSENTOS'] = df_limpo['ASSENTOS'].astype('int64')

df_limpo = df_limpo.drop_duplicates()
```
